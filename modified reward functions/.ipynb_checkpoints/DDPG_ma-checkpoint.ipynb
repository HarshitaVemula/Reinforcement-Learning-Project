```python
class AgentD4PG(ptan.agent.BaseAgent):
    """
    Agent implementing noisy agent
    """
    def __init__(self, net, device="cpu", epsilon=1.0):
        self.net = net
        self.device = device
        self.epsilon = epsilon

    def __call__(self, states, agent_states):
        states_v = ptan.agent.float32_preprocessor(states).to(self.device)
        mu_v = self.net(states_v)
        actions = mu_v.data.cpu().numpy()
        actions += self.epsilon * np.random.normal(size=actions.shape)
        actions = np.clip(actions, -1, 1)
        return actions, agent_states
```

```python
class DDPGActor(nn.Module):
    def __init__(self, obs_size, act_size):
        super(DDPGActor, self).__init__()

        self.net = nn.Sequential(
            nn.Linear(obs_size, 4),
            nn.ReLU(),
            nn.Linear(4, 4),
            nn.ReLU(),
            nn.Linear(4, act_size),
            nn.Tanh()
        )

    def forward(self, x):
        return self.net(x)


class DDPGCritic(nn.Module):
    def __init__(self, obs_size, act_size):
        super(DDPGCritic, self).__init__()

        self.obs_net = nn.Sequential(
            nn.Linear(obs_size, 8),
            nn.BatchNorm1d(8),
            nn.ReLU(),
        )

        self.out_net = nn.Sequential(
            nn.Linear(8 + act_size, 6),
            nn.BatchNorm1d(6),
            nn.ReLU(),
            nn.Linear(6, 1)
        )

    def forward(self, x, a):
        obs = self.obs_net(x)
        return self.out_net(torch.cat([obs, a], dim=1))
```

```python
if __name__ == "__main__":
    N_AGENTS = 2
    GAMMA = 0.99
    BATCH_SIZE = 5000
    LEARNING_RATE_ACTOR = 1e-4
    LEARNING_RATE_CRITIC = 1e-3
    REPLAY_SIZE = 5000
    REPLAY_INITIAL = 100

    TEST_ITERS = 120

    EPSILON_DECAY_LAST_FRAME = 1000
    EPSILON_START = 1.2
    EPSILON_FINAL = 0.02

    device = torch.device("cpu")
    
    for n in range(1):
        
        print(n)
        
        act_net, crt_net, tgt_act_net, tgt_crt_net, agent, exp_source, buffer, act_opt, crt_opt, frame_idx = {}, {}, {}, {}, {}, {}, {}, {}, {}, {}
        rew_last_1000, rew, track_loss_critic, track_loss_actor = {}, {}, {}, {}
        for uid in buildings:
            env[uid].reset()
            #Create as many actor and critic nets as number of agents
            #Actor: states_agent_i -> actions_agent_i
            act_net[uid] = DDPGActor(env[uid].observation_spaces[uid].shape[0], env[uid].action_spaces[uid].shape[0]).to(device)

            #Critic: states_all_agents + actions_all_agents -> Q-value_agent_i [1]
            crt_net[uid] = DDPGCritic(env[uid].observation_spaces[uid].shape[0], env[uid].action_spaces[uid].shape[0]).to(device)

            tgt_act_net[uid] = ptan.agent.TargetNet(act_net[uid])
            tgt_crt_net[uid] = ptan.agent.TargetNet(crt_net[uid])

            agent[uid] = model.AgentD4PG(act_net[uid], device=device)
            exp_source[uid] = ptan.experience.ExperienceSourceFirstLast(env[uid], agent[uid], gamma=GAMMA, steps_count=1)
            buffer[uid] = ptan.experience.ExperienceReplayBuffer(exp_source[uid], buffer_size=REPLAY_SIZE)
            act_opt[uid] = optim.Adam(act_net[uid].parameters(), lr=LEARNING_RATE_ACTOR)
            crt_opt[uid] = optim.Adam(crt_net[uid].parameters(), lr=LEARNING_RATE_CRITIC)

            frame_idx[uid] = 0

            rew_last_1000[uid], rew[uid], track_loss_critic[uid], track_loss_actor[uid] = [], [], [], []

        batch, states_v, actions_v, rewards_v, dones_mask, last_states_v, q_v, last_act_v, q_last_v, q_ref_v, critic_loss_v, cur_actions_v, actor_loss_v = {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}

        while not env[buildings[-1]]._terminal():
            if frame_idx[4]%800 == 0:
                    print(frame_idx[uid])
            for uid in buildings:
                agent[uid].epsilon = max(EPSILON_FINAL, EPSILON_START - frame_idx[uid] / EPSILON_DECAY_LAST_FRAME)
                frame_idx[uid] += 1           
                buffer[uid].populate(1)

            total_electricity_consumption, current_price = 0.0, 0.0
            for uid in buildings:
                total_electricity_consumption = total_electricity_consumption + env[uid].building[uid].current_electricalConsumptionCool
            for uid in buildings:  
                current_price = total_electricity_consumption*3e-5 + 0.045
                electricity_cost = buffer[uid].buffer[-1].reward*current_price
                buffer[uid].buffer[-1] = buffer[uid].buffer[-1]._replace(reward=electricity_cost)
                
            if len(buffer[uid]) < REPLAY_INITIAL:
                continue   
            
            for uid in buildings:
                for k in range(6):
                    batch[uid] = buffer[uid].sample(BATCH_SIZE)
                    states_v[uid], actions_v[uid], rewards_v[uid], dones_mask[uid], last_states_v[uid] = common.unpack_batch_ddqn(batch[uid], device) 


                    # TRAIN CRITIC
                    crt_opt[uid].zero_grad()
                    #Obtaining Q' using critic net with parameters teta_Q'
                    q_v[uid] = crt_net[uid](states_v[uid], actions_v[uid])

                    #Obtaining estimated optimal actions a|teta_mu from target actor net and from s_i+1.
                    last_act_v[uid] = tgt_act_net[uid].target_model(last_states_v[uid]) #<----- Actor to train Critic

                    #Obtaining Q'(s_i+1, a|teta_mu) from critic net Q'
                    q_last_v[uid] = tgt_crt_net[uid].target_model(last_states_v[uid], last_act_v[uid])
                    q_last_v[uid][dones_mask[uid]] = 0.0

                    #Q_target used to train critic net Q'
                    q_ref_v[uid] = rewards_v[uid].unsqueeze(dim=-1) + q_last_v[uid] * GAMMA
                    critic_loss_v[uid] = F.mse_loss(q_v[uid], q_ref_v[uid].detach())
                    critic_loss_v[uid].backward()
                    crt_opt[uid].step()


                    # TRAIN ACTOR
                    act_opt[uid].zero_grad()
                    #Obtaining estimated optimal current actions a|teta_mu from actor net and from s_i
                    cur_actions_v[uid] = act_net[uid](states_v[uid])

                    #Actor loss = mean{ -Q_i'(s_i, a|teta_mu) }
                    actor_loss_v[uid] = -crt_net[uid](states_v[uid], cur_actions_v[uid]) #<----- Critic to train Actor
                    actor_loss_v[uid] = actor_loss_v[uid].mean()
                    #Find gradient of the loss and backpropagate to perform the updates of teta_mu
                    actor_loss_v[uid].backward()
                    act_opt[uid].step()

                    if frame_idx[uid] % 1 == 0:
                        tgt_act_net[uid].alpha_sync(alpha=1 - 0.1)
                        tgt_crt_net[uid].alpha_sync(alpha=1 - 0.1)

#             track_loss_critic[uid].append(critic_loss_v[uid])
#             track_loss_actor[uid].append(actor_loss_v[uid])      
        
        rew = {}
        for uid in buildings:
            rew[uid] = env[uid].building[uid].electricalControlled[3500:6000]
        filename_results = 'DDPG'+str(n)+'.csv'
        results = pd.DataFrame.from_dict(rew)
        results.to_csv(filename_results,index=False)
```

```python
from easydict import EasyDict as edict

torch.manual_seed(7)
if __name__ == "__main__":
    start = time.time()
    
    N_AGENTS = 2
    GAMMA = 0.99
    BATCH_SIZE = 5000
    LEARNING_RATE_ACTOR = 1e-4
    LEARNING_RATE_CRITIC = 1e-3
    REPLAY_SIZE = 5000
    REPLAY_INITIAL = 100

    TEST_ITERS = 120

    EPSILON_DECAY_LAST_FRAME = 1000
    EPSILON_START = 1.2
    EPSILON_FINAL = 0.02


    env.reset()

    device = torch.device("cpu")

    #Create as many actor and critic nets as number of agents
    #Actor: states_agent_i -> actions_agent_i
    act_net = DDPGActor(env.observation_space.shape[0], env.action_space.shape[0]).to(device)

    #Critic: states_all_agents + actions_all_agents -> Q-value_agent_i [1]
    crt_net = DDPGCritic(env.observation_space.shape[0], env.action_space.shape[0]).to(device)

    tgt_act_net = ptan.agent.TargetNet(act_net)
    tgt_crt_net = ptan.agent.TargetNet(crt_net)

    agent = model.AgentD4PG(act_net, device=device)
    exp_source = ptan.experience.ExperienceSourceFirstLast(env, agent, gamma=GAMMA, steps_count=1)
    buffer = ptan.experience.ExperienceReplayBuffer(exp_source, buffer_size=REPLAY_SIZE)
    act_opt = optim.Adam(act_net.parameters(), lr=LEARNING_RATE_ACTOR)
    crt_opt = optim.Adam(crt_net.parameters(), lr=LEARNING_RATE_CRITIC)

    frame_idx = 0
    best_reward = None
    while not env._terminal():
        agent.epsilon = max(EPSILON_FINAL, EPSILON_START - frame_idx / EPSILON_DECAY_LAST_FRAME)
        frame_idx += 1

        buffer.populate(1)
        rewards_steps = exp_source.pop_rewards_steps()
        if rewards_steps:
            rewards, steps = zip(*rewards_steps)

        if len(buffer) < REPLAY_INITIAL:
            continue

        for k in range(10):
            batch = buffer.sample(BATCH_SIZE)
            states_v, actions_v, rewards_v, dones_mask, last_states_v = common.unpack_batch_ddqn(batch, device)

            # TRAIN CRITIC
            crt_opt.zero_grad()
            #Obtaining Q' using critic net with parameters teta_Q'
            q_v = crt_net(states_v, actions_v)

            #Obtaining estimated optimal actions a|teta_mu from target actor net and from s_i+1.
            last_act_v = tgt_act_net.target_model(last_states_v) #<----- Actor to train Critic

            #Obtaining Q'(s_i+1, a|teta_mu) from critic net Q'
            q_last_v = tgt_crt_net.target_model(last_states_v, last_act_v)
            q_last_v[dones_mask] = 0.0

            #Q_target used to train critic net Q'
            q_ref_v = rewards_v.unsqueeze(dim=-1) + q_last_v * GAMMA
            critic_loss_v = F.mse_loss(q_v, q_ref_v.detach())
            critic_loss_v.backward()
            crt_opt.step()


            # TRAIN ACTOR
            act_opt.zero_grad()
            #Obtaining estimated optimal current actions a|teta_mu from actor net and from s_i
            cur_actions_v = act_net(states_v)

            #Actor loss = mean{ -Q_i'(s_i, a|teta_mu) }
            actor_loss_v = -crt_net(states_v, cur_actions_v) #<----- Critic to train Actor
            actor_loss_v = actor_loss_v.mean()
            #Find gradient of the loss and backpropagate to perform the updates of teta_mu
            actor_loss_v.backward()
            act_opt.step()

            if frame_idx % 1 == 0:
                tgt_act_net.alpha_sync(alpha=1 - 0.1)
                tgt_crt_net.alpha_sync(alpha=1 - 0.1)

```

```python
def readSimulationResults(demand_file, weather_file, buildings):
    
    demand = pd.read_csv(r'C:\Users\jrv966\Documents\GitHub\multi-agent-RL\simulationData\AustinResidential_TH.csv',sep="\t")
    weather = pd.read_csv(r'C:\Users\jrv966\Documents\GitHub\multi-agent-RL\simulationData\Austin_Airp_TX-hour.csv',sep="\t")
    weather = weather.rename(columns = {' Ta':'Ta',' h':'h'})

    #Obtaining all the building IDs from the header of the document with the results
    ids = []
    for col_name in demand.columns:
        if "(" in col_name:
            ids.append(col_name.split('(')[0])

    #Creating a list with unique values of the IDs
    unique_id_str = sorted(list(set(ids)))
    unique_id = [int(i) for i in unique_id_str if int(i) in buildings]
    unique_id_file = ["(" + str(uid) for uid in unique_id]

    return demand, weather, unique_id, unique_id_file

class DemandResponseEnvironmentCoolingMA(gym.Env):  
    def __init__(self, demand_file, weather_file, buildings_sim = [], time_resolution = 1, simulation_period = (3500,6000)):
        class Building:  
            def __init__(self, buildingId):
                class HeatPump:
                    #Variables
                    self.nominalPower = 0
                    self.maxCooling = []
                    self.maxHeating = []
                    self.copHeating = []
                    self.copCooling = []
                    self.electricalConsumptionCool = []
                    self.electricalConsumptionHeat = []
                    self.heatSupply = []
                    self.coolingSupply = []
                    self.tTargetHeating = []
                    self.tTargetCooling = []
                    self.etaTech = []
                class HeatTank:
                    self.capacity = []
                    self.heatStored = []
                    self.energyBalance = []

                class CoolTank:
                    self.capacity = []
                    self.coolingStored = []
                    self.energyBalance = []

                self.buildingId = buildingId
                self.sim_results = {} #'cooling_demand','heating_demand','non_shiftable_load','t_in','t_out','hour'
                self.electricalControlled = []

                #Devices
                self.heatPump = HeatPump()
                self.heatTank = HeatTank()
                self.coolTank = CoolTank()
                
        self.time_resolution = time_resolution
        self.n_buildings = len(buildings_sim)   
        self.simulation_period = simulation_period
        self.a_track = {}
        #Creating data frames from the results of the simulation in CitySim
        demand, weather, self.unique_id, unique_id_file = readSimulationResults(demand_file, weather_file, buildings_sim)

        #Defining state space: hour, Tout, Tin, Thermal_energy_stored
        self.high_state = {self.unique_id[i]: np.array([24.0, 40.0, 1.001]) for i in range(self.n_buildings)}
        self.low_state = {self.unique_id[i]: np.array([1.0, 17.0, -0.001]) for i in range(self.n_buildings)}
        self.observation_spaces = {self.unique_id[i]: spaces.Box(low=self.low_state[self.unique_id[i]], high=self.high_state[self.unique_id[i]], dtype=np.float32) for i in range(self.n_buildings)}
        
        #Defining action space: new desired energy stored in the tank
        self.min_action = {self.unique_id[i]: np.array([-0.5]) for i in range(self.n_buildings)}
        self.max_action = {self.unique_id[i]: np.array([0.5]) for i in range(self.n_buildings)}
        self.action_spaces = {self.unique_id[i]: spaces.Box(low=self.min_action[self.unique_id[i]], high=self.max_action[self.unique_id[i]], dtype=np.float32) for i in range(self.n_buildings)}
        
        #Filling out the variables of the buildings with the output files from CitySim
        self.building = {}
        for uid, building_id in zip(self.unique_id, unique_id_file):
            self.a_track[uid] = []
            self.building[uid] = Building(uid)
  
            sim_var_name = {sim_var: [col for col in demand.columns if building_id in col if sim_var in col] for sim_var in list(['Qs','ElectricConsumption','Ta'])}

            #Summing up and averaging the values by the number of floors of the building
            qs = demand[sim_var_name['Qs']]
            self.building[uid].sim_results['cooling_demand'] = -qs[qs<0].sum(axis = 1)/1000
            self.building[uid].sim_results['heating_demand'] = qs[qs>0].sum(axis = 1)/1000
            self.building[uid].sim_results['non_shiftable_load'] = demand[sim_var_name['ElectricConsumption']]
            self.building[uid].sim_results['t_in'] = demand[sim_var_name['Ta']].mean(axis = 1)
            self.building[uid].sim_results['t_out'] = weather['Ta']
            self.building[uid].sim_results['hour'] = weather['h']

            # HEAT PUMPS
            self.building[uid].heatPump.tTargetHeating = 45+273.150001
            self.building[uid].heatPump.tTargetCooling = 10+273.150001
            self.building[uid].heatPump.etaTech = 0.22

            #Calculating COPs of the heat pumps for every hour
            self.building[uid].heatPump.copHeating = self.building[uid].heatPump.etaTech*self.building[uid].heatPump.tTargetHeating/(self.building[uid].heatPump.tTargetHeating - (self.building[uid].sim_results['t_out'] + 273.15))
            self.building[uid].heatPump.copHeating[self.building[uid].heatPump.copHeating < 0] = 20.0
            self.building[uid].heatPump.copHeating[self.building[uid].heatPump.copHeating > 20] = 20.0
            self.building[uid].heatPump.copCooling = self.building[uid].heatPump.etaTech*self.building[uid].heatPump.tTargetCooling/(self.building[uid].sim_results['t_out'] + 273.15 - self.building[uid].heatPump.tTargetCooling)
            self.building[uid].heatPump.copCooling[self.building[uid].heatPump.copCooling < 0] = 20.0
            self.building[uid].heatPump.copCooling[self.building[uid].heatPump.copCooling > 20] = 20.0

            #We assume that the heat pump is large enough to meet the highest heating or cooling demand of the building (Tindoor always satisfies Tsetpoint)
            self.building[uid].heatPump.nominalPower = max(max(self.building[uid].sim_results['cooling_demand']/(self.building[uid].heatPump.copCooling-0.00001)),max(self.building[uid].sim_results['heating_demand']/(self.building[uid].heatPump.copHeating+0.00001)))

            #Defining the capacity of the heat tanks based on the average non-zero heat demand
            self.building[uid].heatTank.capacity = max(self.building[uid].sim_results['heating_demand']/(self.building[uid].heatPump.copHeating+0.00001))*3
            self.building[uid].coolTank.capacity = max(self.building[uid].sim_results['cooling_demand']/(self.building[uid].heatPump.copCooling-0.00001))*7
        del demand, weather, qs
        #Initializing the variables
        self.reset()        

    def step(self, action):
        action = action/self.time_resolution
        #Ref: Assessment of energy efficiency in electric storage water heaters (2008 Energy and Buildings)
        heatLossFactor = 1 - 0.19/24
        power_cap, eps, reward = 0.4, 0.000001, 0.0
        
        if self.current_building == self.unique_id[-1]:
            self.agent = iter(np.array(self.unique_id))
            self.time_step = [next(self.hour) for j in range(self.time_resolution)][-1]
            
        self.current_building = next(self.agent)
        uid = self.current_building   
        
        self.a_track[uid].append(action)
        
        reward = 0
        for i in range(self.time_step, self.time_step + self.time_resolution):                
            #Heating
            heat_power_avail = self.building[uid].heatPump.maxHeating[i] - self.building[uid].sim_results['heating_demand'][i]
            delta = max(-self.building[uid].heatTank.heatStored[i],min(0*self.building[uid].heatTank.capacity, self.building[uid].heatTank.capacity - self.building[uid].heatTank.heatStored[i]))
            self.building[uid].heatPump.heatSupply[i] = max(0,self.building[uid].sim_results['heating_demand'][i]+min(power_cap*heat_power_avail,delta))
            self.building[uid].heatPump.electricalConsumptionHeat[i] = self.building[uid].heatPump.heatSupply[i]/(self.building[uid].heatPump.copHeating[i]+eps)
            self.building[uid].heatTank.energyBalance[i] = self.building[uid].heatPump.electricalConsumptionHeat[i]*(self.building[uid].heatPump.copHeating[i]+eps) - self.building[uid].sim_results['heating_demand'][i]
            self.building[uid].heatTank.heatStored[i] = self.building[uid].heatTank.heatStored[i] + self.building[uid].heatTank.energyBalance[i]

            #Cooling:
            heat_power_avail = self.building[uid].heatPump.maxCooling[i] - self.building[uid].sim_results['cooling_demand'][i]
            delta = max(-self.building[uid].coolTank.coolingStored[i],min(action*self.building[uid].coolTank.capacity, self.building[uid].coolTank.capacity - self.building[uid].coolTank.coolingStored[i]))
            self.building[uid].heatPump.coolingSupply[i] = max(0,self.building[uid].sim_results['cooling_demand'][i]+min(power_cap*heat_power_avail,delta))
            self.building[uid].heatPump.electricalConsumptionCool[i] = self.building[uid].heatPump.coolingSupply[i]/(self.building[uid].heatPump.copCooling[i]+eps)
            self.building[uid].coolTank.energyBalance[i] = self.building[uid].heatPump.electricalConsumptionCool[i]*(self.building[uid].heatPump.copCooling[i]+eps) - self.building[uid].sim_results['cooling_demand'][i]
            self.building[uid].coolTank.coolingStored[i] = self.building[uid].coolTank.coolingStored[i] + self.building[uid].coolTank.energyBalance[i]

            if i < self.simulation_period[1]:
                self.building[uid].heatTank.heatStored[i+1] = heatLossFactor*self.building[uid].heatTank.heatStored[i]
                self.building[uid].coolTank.coolingStored[i+1] = heatLossFactor*self.building[uid].coolTank.coolingStored[i]

            self.building[uid].electricalControlled[i] = self.building[uid].heatPump.electricalConsumptionHeat[i] + self.building[uid].heatPump.electricalConsumptionCool[i]
            self.total_electrical_consumption[i] = self.total_electrical_consumption[i] + self.building[uid].heatPump.electricalConsumptionCool[i]
            self.building[uid].current_electricalConsumptionCool = self.building[uid].heatPump.electricalConsumptionCool[i]
            reward = reward - self.building[uid].heatPump.electricalConsumptionCool[i]
            
            if self.current_building == self.unique_id[-1]:
                self.electricity_price[i] = self.total_electrical_consumption[i]*3e-5 + 0.045
                self.current_price = self.electricity_price[i]
                
            if (i+1 > self.simulation_period[1]) and (self.current_building == self.unique_id[-1]): break
            
        #States: hour, Tout, Tin, Thermal_energy_stored        
        cooling_stored = self.building[uid].coolTank.coolingStored[i]/self.building[uid].coolTank.capacity

        #States: hour, Tout, Tin, Thermal_energy_stored
        self.state = np.array([self.building[uid].sim_results['hour'][i], self.building[uid].sim_results['t_out'][i], cooling_stored],dtype=np.float32)

        terminal = self._terminal()
        
        return (self.state, reward, terminal, {})
    
    def reset(self):
        #Initialization of variables
        self.hour = iter(np.array(range(self.simulation_period[0], self.simulation_period[1] + 1)))
        self.time_step = next(self.hour)
        self.agent = iter(np.array(self.unique_id))
        self.current_building = 0
        self.state = np.array([0.0,0.0,0.0], dtype=np.float32)
        self.building_ids = []
        var_length = self.simulation_period[1] + 1
        for uid in self.unique_id:
            #Initialization of the heat and cool tanks (empty)
            self.building[uid].heatTank.heatStored = np.zeros(var_length)
            self.building[uid].coolTank.coolingStored = np.zeros(var_length)
            self.building[uid].heatPump.electricalConsumptionHeat = np.zeros(var_length)
            self.building[uid].heatPump.electricalConsumptionCool = np.zeros(var_length)
            self.building[uid].heatPump.heatSupply = np.zeros(var_length)
            self.building[uid].heatPump.coolingSupply = np.zeros(var_length)
            self.building[uid].heatTank.energyBalance = np.zeros(var_length)
            self.building[uid].coolTank.energyBalance = np.zeros(var_length)
            self.building[uid].electricalControlled = np.zeros(var_length)
            self.building[uid].heatPump.maxHeating = self.building[uid].heatPump.nominalPower*self.building[uid].heatPump.copHeating
            self.building[uid].heatPump.maxCooling = self.building[uid].heatPump.nominalPower*self.building[uid].heatPump.copCooling
        
        self.current_price = 0.0
        self.building[uid].current_electricalConsumptionCool = 0.0
        self.electricity_price = np.zeros(var_length)
        self.total_electrical_consumption = np.zeros(var_length)
                
        return self._get_ob()
    
    def _get_ob(self):
        s = self.state
        return np.array([s[0], s[1], s[2]], dtype=np.float32)
    
    def _terminal(self):
        return bool(self.time_step >= self.simulation_period[1] and self.current_building == self.unique_id[-1])
    
    def seed(self, seed=None):
        self.np_random, seed = seeding.np_random(seed)
        return [seed]
```
